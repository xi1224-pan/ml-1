```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 1. 載入 Hugging Face 線上模型
model_name = "uer/roberta-base-finetuned-jd-binary-chinese"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# 2. 定義情緒分析函數（回傳情緒 + 分數）
def predict_sentiment_with_score(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=64)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)
        score = probs[0][1].item() - probs[0][0].item()  # 正面機率 - 負面機率

    # 分類規則：你可以根據需求調整
    if score > 1:
        sentiment = "強烈正面"
    elif score > 0:
        sentiment = "正面"
    elif score == 0:
        sentiment = "中立"
    else:
        sentiment = "負面"

    return {
        "情緒分類": sentiment,
        "情緒分數": round(score, 4),

    }

# 3. 測試範例
print(predict_sentiment_with_score("跟小朋友一起絕對是首推吧，小朋友選耀希就無敵了，加上，又沒時間限制可以慢慢通關"))
print(predict_sentiment_with_score("股市下跌"))
print(predict_sentiment_with_score("先防個雷加前言好了距離破關大概已經一個多月了，當初是趕著薩爾達中文化前的一個月開始趕進度，然後就沈浸在海拉魯的世界，直到現在才打心得。當初有接觸過的瑪利歐有家一三四代，還有gameboy上的二代，這上面都不是自己的，所以說實在也沒全破過。直到前年底iphone出了一款瑪利歐RUN才算是第一款購買的。因為沒玩過3D的瑪利歐，相信有不少以前不常碰遊戲的人會認為， 瑪利歐3D的會好玩嗎？打從奧德賽未上市前，在看預告的時候仍然保持這樣的感覺。直到開啟遊戲後，進入瀑布國，那漂亮的場景還有預告片所聽到的熟悉音樂就完全改觀了，原來附身如此多變化、變成鐵球撞瀑布國兔王的狗狗尋找月亮、紫幣是這麼的有樂趣，還有打完頭目後有更多的月亮隱藏在各式各樣場景及小關卡裡面，可以說在破關前，是非常不容易玩到膩的！部份關卡融入了8bit風格的遊戲內容，也是個人相當喜歡的一部份，補足了以前只能玩一下下的遺憾，尤其都市國的慶典及裡面的1-1小關卡，連隱藏的1UP都有保留，真的會讓人對這款遊戲愛不釋手。月之國更背面的超長型關卡，對喜歡動作遊戲的應該也都會很滿足才是，真的充滿了挑戰啊！一段一段的摸熟，然後慢慢推進，真的覺得很不容易，但突破之後就會覺得其實也沒那麼難嘛！接下來說說一些個人覺得比較困難或有創意的紫幣及月亮紫幣的話，除了料理國有3顆沒找到看YOUTUBE外，其他都靠一己之力，結果看完後那三顆居然就在起點而已……我真的逛超久。月亮的話，大家常提的月之背面的兩個無帽關，無帽關拿到之後才發現，原來有一模一樣的關卡但是有帽子，這邊當然就輕而易舉的拿到另外一個就是小遊戲中最高等級的彈跳博爾吧，個人覺得這比無帽關還難…跳繩反而後面不看螢幕照節奏按很容易，不到十次就一百下了。其他有一些是被雷到的月亮像是吐金幣種花…說真的沒看到板友推文我也想不到。最覺得討厭的是看照片找月亮，像大家來找碴這種明顯的還算簡單，那個都市國的keep…能想到的我真的覺得很神，一些完全想不到的我是直接放棄到網路上看攻略（畢竟要去救薩爾達沒時間再耗了）總之，有入手 Switch的真的非常推薦同時入手這款啊！即使是3D的瑪利歐，玩完後百分之八九十不會後悔的！"))
```
